```python
# Modules
import pandas as pd
import numpy as np
import os
import glob
import shutil
```

```python
# Get Songs into Dataframe
songs_df = pd.read_csv('all_songs.csv')

# add album path
songs_df["album"] = [k.split('/')[0] for k in songs_df.path]

albums_df = pd.DataFrame(songs_df.album.drop_duplicates())
albums_df.columns = ['album']

songs_df
```

```python
# Make folders for each album

for ind in albums_df.index:
    album = albums_df['album'][ind]
    album_path = os.path.join('./videos/', album)

    songs_in_album_df = songs_df[songs_df.album == album]

    album_description = album

    for ind in songs_in_album_df.index:
        song_name = songs_in_album_df['path'][ind].split('/')[1].replace(".mp3","")
        youtube_link = songs_in_album_df['youtube_link'][ind]
        album_description = album_description + "\n" + song_name + " - " + youtube_link

    with open('./album-descriptions/' + album + ".txt", 'w') as f:
        f.write(album_description)
```
